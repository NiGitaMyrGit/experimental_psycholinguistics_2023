```python
import seaborn as sns
import jsonlines
import pandas as pd
import os
```

```python
raw = os.path.join("..", "in", "jatos_results_20230920125013.txt")
```

```python
i = 0

with jsonlines.open(raw) as reader:
    for line in reader:
        if i == 0:
            df = pd.DataFrame(line)
            i += 1
        else:
            df = pd.concat([df, pd.DataFrame(line)])
            i += 1
```
