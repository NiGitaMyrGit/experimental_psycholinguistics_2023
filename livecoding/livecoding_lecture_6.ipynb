```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# n= amount of times the coin should be flipped
# p=1/2 flip a coing, make it eaither a 1 or a 2

list(np.random.binomial(n=10, p=1/2, size=10))
```

```python
# the more times we flip the coins, the more the graph start to resemble the 50/50 chance we are expecting
# The more data, the finer the distribution
# change p to 1/6 and you get a die, maybe do the same for a 20 sided die
flips = np.random.binomial(n=100, p=1/20, size=100000)
ax = sns.histplot(flips)
ax.set(xlabel="number of experiments",
       ylabel= "number of heads")
plt.xlim(0,100)
#plt.xticks(range(0,11,1))
sns.despine()
```
