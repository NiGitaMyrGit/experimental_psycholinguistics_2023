### a-responses

```python
json_files = glob.glob(os.path.join("in","a","*.json"))  # Adjust the path to your actual directory
dataframes = [pd.read_json(file) for file in json_files]
a_df = pd.concat(dataframes, ignore_index=True)
a_df
```

```python
useful_col = ["sound", "response"]
df_a = a_df[useful_col].copy()
df_a['sound'] = df_a['sound'].astype(str)
df_a 
```

```python
correct_df = pd.read_csv(os.path.join("in","sentences_3.csv"), delimiter=";", decimal=",")
correct_df
```

```python
merged_df_a = pd.merge(df_a, correct_df, on='sound')
merged_df_a = merged_df_a.sort_values(by='sound', ascending=True)
#df_a_sorted
merged_df_a
```

```python
# Convert "response" and "sentences" columns to lowercase
merged_df_a['response'] = merged_df_a['response'].str.lower()
merged_df_a['sentences'] = merged_df_a['sentences'].str.lower()

# tokenize the words 
merged_df_a['response_words'] = merged_df_a['response'].apply(word_tokenize)
merged_df_a['sentence_words'] = merged_df_a['sentences'].apply(word_tokenize)

# Initialize spell checker
spell = SpellChecker()

# Correct spelling in response words
merged_df_a['response_words'] = merged_df_a['response_words'].apply(lambda words: [spell.correction(word) for word in words])


# Comparing the lists of words and counting the correct ones
merged_df_a['correct_word_count'] = [sum(word in sentence_words for word in response_words)
                                   for response_words, sentence_words
                                   in zip(merged_df_a['response_words'], merged_df_a['sentence_words'])]

print(merged_df_a[['sound', 'response', 'sentences', 'correct_word_count']])
```

```python
# Perform t-test using pingouin
t_test_result = pg.ttest(merged_df_a['correct_word_count_x'][merged_df_a['sound'].str.startswith('n-')],
                         merged_df_a['correct_word_count_x'][~merged_df_a['sound'].str.startswith('n-')],
                         paired=False)

# Display t-test result
print(t_test_result)
```

```python
# Save as csv-file
merged_df_a.to_csv(os.path.join('out','df_a.csv'), index=False)
```
